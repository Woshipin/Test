```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('coffee.csv')
df

# Group by 'category' and 'product', then calculate the total orders for each product type
total_orders = df.groupby(['product_category', 'product_type']).agg(total_order=('transaction_qty', 'sum')).reset_index()

# Display the resulting DataFrame
print(total_orders)

# Save the results to a CSV file
output_file_path = 'total_orders_by_product.csv'
total_orders.to_csv(output_file_path, index=False)

# Plot to visualize the total orders by product
plt.figure(figsize=(12,10))
plt.barh(total_orders['product_type'], total_orders['total_order'], color='skyblue')
plt.xlabel('Total Orders')
plt.ylabel('Product')
plt.title('Total Orders for Each Product Type')
plt.show()
```
